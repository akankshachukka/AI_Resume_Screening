```python
#3. Step 1 — Load Resumes
import pdfplumber
import os

def read_pdf(file_path):
    text = ""
    with pdfplumber.open(file_path) as pdf:
        for page in pdf.pages:
            text += page.extract_text()
    return text


resumes = []

file_path= "C:\\Users\\Akanksha Chukka\\Downloads\\AkankshaC.pdf"
folder = "resumes"

for file in os.listdir(folder):
    if file.endswith(".pdf"):
        text = read_pdf(os.path.join(folder, file))
        resumes.append(text)


print(len(resumes))
```

```python
#Step 2 — Load Job Description
with open(r"C:\Resume_Project\Job Description.txt", "r") as f:
    job_desc = f.read()
```

```python
#Step 3 — Use BERT for Embeddings
from sentence_transformers import SentenceTransformer

model = SentenceTransformer("all-MiniLM-L6-v2")

resume_embeddings = model.encode(resumes)
job_embedding = model.encode([job_desc])
```

```python
#Step 4 — Calculate Similarity Score
from sklearn.metrics.pairwise import cosine_similarity

scores = cosine_similarity(job_embedding, resume_embeddings)

print(scores)
```

```python
#Step 5 — Rank Resumes
results = []

for i, score in enumerate(scores[0]):
    results.append((i, score))

results = sorted(results, key=lambda x: x[1], reverse=True)

print("Ranking:")

for r in results:
    print(r)
```
